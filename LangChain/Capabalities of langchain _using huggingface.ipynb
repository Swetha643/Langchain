```python
from langchain import HuggingFaceHub

llm=HuggingFaceHub(repo_id="stabilityai/stablecode-completion-alpha-3b-4k")
prompt="import torch\nimport torch.nn as nn"
completion=llm(prompt)

print(completion)
```

**Setting the Prompt:**

prompt = "import torch\nimport torch.nn as nn":

 This defines the prompt to be fed to the LLM. It consists of two lines of Python code that import necessary modules for deep learning:

import torch: Imports the PyTorch library, a popular framework for building neural networks.

import torch.nn as nn: Imports the nn module from PyTorch, which provides building blocks for constructing neural networks.

2. **Prompts**

```python
from langchain import PromptTemplate
template="Name a market leader that makes {product}"

prompt=PromptTemplate(

      input_variables=['product'],
      template=template
)

prompt.format(product="computers")
```

```python

email_template=""" Write an email to your {receiver_name} starting about your reason of absence from office for {x} days.
The email should not exceed 50 words and must contain a subject"""


prompt2=PromptTemplate(

      input_variables=['receiver_name','x'],
      template=email_template
)

prompt2.format(receiver_name="HR",x=5)
```

A good prompt is what makes your language models really useful. With the prompts templates you can organize and created standard prompt templates for generating information repeatedly.

3. **Chains**

```python
from langchain import LLMChain


llm2=HuggingFaceHub(repo_id="openai-community/gpt2-large")
chain=LLMChain(llm=llm2,prompt=prompt2)

chain.run(receiver_name="HR", x=3)
```

4. **Memory**

```python
from langchain import ConversationChain
Conversation=ConversationChain(llm=llm2, verbose=True)
Conversation.predict(input="How r u")
Conversation.predict(input="The learners have been very supportive.")

```
